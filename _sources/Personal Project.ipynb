### B. Pemeriksaan LOCAL OUTLIER FACTOR (LOF)

Langkah berikutnya adalah deteksi outlier. Deteksi outlier adalah metode untuk mengidentifikasi titik-titik data yang dianggap sebagai outlier atau anomali. Outlier adalah nilai yang secara signifikan berbeda dari sebagian besar data lainnya dalam sebuah dataset. Mengidentifikasi outlier bermanfaat untuk meningkatkan akurasi analisis dan model prediktif. 

### PERHITUNGAN  LOCAL OUTLIER FACTOR (LOF) MENGGUNAKAN PYTHON

Untuk perhitungan LOF menggunakan Python ini dapat  menggunakan library scikit-learn. Cara ini mempermudah dan mempercepat proses deteksi outlier. Berikut adalah kode untuk menghitung LOF:

```
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
import numpy as np

# Baca dataset
df = pd.read_csv('/work/Vertebral_Dataset.csv')

# Siapkan data untuk LOF (drop kolom 'Class' karena fitur kategorikal)
x = df.drop('Class_att', axis=1)

# Inisialisasi model LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.040)

# Terapkan LOF pada dataset
y_pred = lof.fit_predict(x) # fit_predict(x): Menerapkan model LOF pada dataset x dan mengembalikan prediksi. Hasil prediksi berupa label -1 untuk outlier dan 1 untuk inlier.
# Skor anomali (semakin rendah, semakin besar kemungkinan outlier)
anomaly_scores = lof.negative_outlier_factor_ # negative_outlier_factor_: Menghasilkan skor anomali untuk setiap sampel dalam dataset. Semakin rendah nilainya, semakin besar kemungkinan sampel tersebut adalah outlier.
# Tambahkan kolom hasil prediksi dan skor anomali ke dalam dataframe
df['LOF_Prediksi'] = y_pred # Menambahkan kolom LOF_Prediksi ke DataFrame df yang berisi hasil prediksi LOF (1 untuk inlier, -1 untuk outlier).
df['LOF_Skor_Anomali'] = anomaly_scores # Menambahkan kolom LOF_Skor_Anomali yang berisi skor anomali dari LOF.

# Visualisasi hasil (opsional)
plt.figure(figsize=(10, 6)) # Membuat figure baru dengan ukuran 10x6.
plt.title("LOF Anomaly Scores") # Memberi judul pada plot.
plt.scatter(np.arange(len(anomaly_scores)), anomaly_scores, c=(y_pred == -1), cmap='coolwarm', edgecolor='k') # Membuat scatter plot dari skor anomali.
plt.axhline(y=-1.5, color='r', linestyle='--')  # Menambahkan garis horizontal pada y = -1.3 untuk memisahkan outlier (opsional, dapat disesuaikan dengan konteks data).
plt.xlabel("Indeks")
plt.ylabel("Skor Anomali LOF")
plt.show()
```

# ENSEMBLE LEARNING

Setelah dilakukan Pemodelan, ada banyak cara untuk meningkatkan akurasi prediksi dari model yang telah kita buat salah satunya adalah Ensemble Learning.

    Ensemble learning adalah teknik yang digunakan untuk meningkatkan kinerja dan akurasi model-model pembelajaran mesin dengan menggabungkan hasil prediksi dari beberapa model atau algoritma yang berbeda.

    Istilah ensemble merujuk pada konsep menggabungkan banyak elemen menjadi satu kesatuan. Konsep penggabungan berbagai pendekatan pembelajaran mesin diharapkan dapat mengatasi kelemahan individu dari masing-masing model dan menciptakan prediksi yang lebih akurat dan stabil.

    Cara kerja dari metode ensemble melibatkan penggunaan beberapa model yang beragam dalam satu tim. Hasil prediksi dari masing-masing model tersebut digabungkan untuk membuat keputusan akhir.

    Pada project kali ini saya akan menggunakan 2 metode dalam ensemble learning yaitu Stacking dan Bagging dan mencari metode mana yang terbaik untuk data kita.

## METODE STACKING

### 1. Pengertian Metode Stacking

 Stacking adalah sebuah metode dimana sebuah model ditraining untuk menggabungkan beberapa individual model yang disebut first-level learner, sedangkan yang digabungkan disebut second-level learner atau meta-learner.

    Stacking melibatkan dua atau lebih base learner atau model dasar sebagai model level-1, dan meta-learner yang menggabungkan prediksi base learner disebut sebagai level-2. Model yang berbeda digunakan oleh base learner untuk belajar dari suatu dataset. Output dari masing-masing model base learner dikumpulkan untuk membuat dataset baru. Di dalam dataset yang baru, setiap sampel berhubungan dengan nilai pada dataset awal yang seharusnya diprediksi. Selanjutnya dataset tersebut digunakan oleh model meta-learner hingga mendapatkan hasil akhir.

### PERCOBAAN METODE STACKING MENGGUNAKAN 2 MODEL FIRST LEARNER DENGAN KNN (N-3) dan KNN (N-5)

1) Pertama, mengambil traning set dari pemisahan data menjadi beberapa variable sebelumnya yang bernama variable X_train (sebagai data training dari fitur) dan y_train (sebagai data training dari kategori)
2) Lalu kita nyatakan sebuah model yang KNN dengan n = 5 dan n = 3, setelah itu berarti kita telah mempunyai 2 first - level learner **bagian 2**
3) Langkah selanjutnya yaitu fit (latih) model tersebut menggunakan data training (**bagian 3**)
4) Langkah selanjutnya yaitu lakukan prediksi terhadap data training tersebut sehingga menghasilkan training data set baru yang dimana gabungan antara hasil prediksi dari first learner dan kategori dari data training sebelumnya (y_train), Prediksi bisa dilihat pada **bagian 4** dan penggabungan data dan encoding label agar inputan berupa numeric pada **bagian 5**

Maka akan didapatkan data set baru hasil pembelajaran kedua First Learner yang kita gunakan untuk menghitung menggunakan meta model

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd 

# Awal Bagian 2
clf_KNN3= KNeighborsClassifier(n_neighbors=3)
clf_KNN5 = KNeighborsClassifier(n_neighbors=5)
# Akhir bagian 2

#awal bagian 3
clf_KNN3.fit(X_train, y_train)
clf_KNN5.fit(X_train, y_train)
#akhir bagian 3

#awal bagian 4
y_pred_KNN3 = clf_KNN3.predict(X_train)
y_pred_KNN5 = clf_KNN5.predict(X_train)
#akhir bagian 4

#awal bagian 5
le = LabelEncoder()

df_pred_KNN3_enc = le.fit_transform(y_pred_KNN3)
df_pred_KNN5_enc = le.fit_transform(y_pred_KNN5) # proses encoding


df_pred_KNN3 = pd.DataFrame(df_pred_KNN3_enc, columns=["KNN3"])
df_pred_KNN5 = pd.DataFrame(df_pred_KNN5_enc, columns=["KNN5"]) # penggabungan data fisrt learner

df_train = pd.DataFrame(y_train)

y_train = y_train.reset_index()
y_train = y_train.drop(columns=['index'])


df_pred_combined_train= pd.concat([df_pred_KNN3, df_pred_KNN5, y_train], axis=1) #penggabungan data  dengan y_train
df_pred_combined_train # df_pred_combined merupakan dataset baru hasil dari kedua model first learner

#akhir bagian 5
```

## METODE BAGGING

Bagging (bootstrap aggregating) merupakan metode ensemble yang melibatkan pelatihan beberapa model secara sendiri pada dataset yang diacak, dan menggabungkan prediksi mereka melalui pemungutan suara.

    setiap model dilatih pada dataset yang diacak dari data yang diambil sampelnya,  data individual dapat dipilih lebih dari satu kali. data acak ini dikenal sebagai sampel bootstrap. Dengan melatih model pada bootstrap yang berbeda, bagging mengurangi varians dari masing-masing model. Bagging juga dapat menghindari overfitting.

    Prediksi dari semua model yang diambil sampelnya kemudian digabungkan dan dibuat voting yang dimana hasil prediksi terbanyak akan di keluarkan sebagai hasil prediksi akhir

    Bagging sangat efektif dalam mengurangi varians dan overfitting, membuat model lebih tangguh dan akurat, terutama dalam kasus di mana masing-masing model rentan terhadap variabilitas tinggi.

### MELAKUKAN PREDIKSI DENGAN METODE BAGGING

- Pertama,fungsi akan dibuat dengan nama bagging dan mengambil parameter x_train, y_train,x_test,y_test,jumlah_juri 
- Kedua, melakukan perulangan yang dimulai dari i = 1 sampai jumlah_juri
- Ketiga, pada perulangan fungsi BaggingClassifier dari sklearn akan dipanggil dengan menginputkan beberapa paramter Yang pertama **base_estimator** berfungsi untuk memberitahu metode apa yang ingin kita gunakan kali ini kita akan memakai GaussianNB(), **n_estimator** berfungsi untuk memberitahu berapa banyak juri yang akan kita gunakan, pada paramater ini inputnya dinamis mengikuti nilai dari i. 
- Keempat, lakukan fit (latih) dengan x_trian dan y_train dari inputan parameter kita sebelumnya
- Kelima, lakukan prediksi x_test dengan model yang kita latih sebelumnya sesuai dengan n_estimator
- Keenam, Lalu hasil dari prediksi kita Encoder ke tipe numeric agar bisa kita hitung akurasinya
- Ketujuh, memasukkan data ke array accuracy yang kita akan gunakan sebagai pengelompokan accuracy dari beberapa n_estimator
- Kedelapan, memilih tingkat akurasi tertinggi dan jumlah n_estimator terendah

pada percobaan kali ini saya menemukan akurasi tertinggi sebesar 83.25% dengan n_estimator = 2

```
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
import pickle


def bagging(x_train, y_train,x_test,y_test,jumlah_juri):
    accuracy = []
    # menentukan rentang mulai dari 1 hingga jumlah_juri (termasuk) dengan langkah penambahan sebesar 1 pada setiap iterasi.
    for i in range(1,jumlah_juri + 1, 1): 
        clf = BaggingClassifier(base_estimator=GaussianNB(),n_estimators=i, random_state=0).fit(x_train, y_train)
        y_pred = clf.predict(x_test)
        le = LabelEncoder()
        y_pred_enc = le.fit_transform(y_pred)
        acc = accuracy_score(y_test, y_pred_enc)
        accuracy.append(acc)
    return accuracy


jumlah_juri = 10
y = bagging(x_train,y_train,x_test,y_test_enc,jumlah_juri)
y

df = pd.DataFrame({
    'n_estimator' : range(1, jumlah_juri + 1, 1),
    'Accuracy' : y
})
df
```
